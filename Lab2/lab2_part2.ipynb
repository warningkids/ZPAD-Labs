### Завдання 1

Завантажити та відкрити датасет Individual Household Electric Power Consumption Dataset.

```python
def load_dataset():

    df = pd.read_csv(
        "household_power_consumption.txt",
        sep=";",
        low_memory=False
    )

    print("Датасет успішно завантажено")
    print("Кількість записів:", len(df))

    return df
```

### Завдання 2

Здійснити очищення даних (Data Cleaning):
- обробити пропущені значення
- привести типи даних до числових

```python
def clean_data(df):

    # заміна ? на NaN
    df.replace("?", np.nan, inplace=True)

    # видалення пропусків
    df.dropna(inplace=True)

    # переведення колонок у числовий формат
    numeric_columns = [
        "Global_active_power",
        "Global_reactive_power",
        "Voltage",
        "Global_intensity",
        "Sub_metering_1",
        "Sub_metering_2",
        "Sub_metering_3"
    ]

    for col in numeric_columns:
        df[col] = df[col].astype(float)

    print("Data cleaning завершено")
    print("Кількість записів після очищення:", len(df))

    return df
```

```python
df = clean_data(df)
```

### Завдання 3

Обрати всі записи, у яких загальна активна споживана потужність перевищує 5 кВт.

```python
def power_over_5kw(df):

    result = df[df["Global_active_power"] > 5]

    print("Кількість записів:", len(result))

    return result
```

```python
power_over_5kw(df)
```

### Завдання 4

Обрати всі записи, у яких сила струму лежить в межах 19–20 А.
Для них визначити ті, у яких пральна машина та холодильник
споживають більше електроенергії ніж бойлер та кондиціонер.

```python
def current_range(df):

    filtered = df[
        (df["Global_intensity"] >= 19) &
        (df["Global_intensity"] <= 20)
    ]

    result = filtered[
        (filtered["Sub_metering_1"] + filtered["Sub_metering_2"]) >
        filtered["Sub_metering_3"]
    ]

    print("Кількість записів:", len(result))

    return result
```

```python
current_range(df)
```

### Завдання 5

Обрати випадковим чином 500000 записів без повторів
та обчислити середні значення трьох груп споживання електроенергії.

```python
def random_sample_mean(df):

    sample = df.sample(n=500000, replace=False)

    means = sample[[
        "Sub_metering_1",
        "Sub_metering_2",
        "Sub_metering_3"
    ]].mean()

    print("Середні значення:")
    print(means)

    return means
```

```python
random_sample_mean(df)
```

### Завдання 6

Обрати записи після 18:00 із середнім споживанням понад 6 кВт,
визначити ті, де група 2 має найбільше споживання.
Після цього обрати кожен третій результат із першої половини
та кожен четвертий із другої половини.

```python
def complex_selection(df):

    # перетворення часу
    df["Time"] = pd.to_datetime(df["Time"])

    evening = df[df["Time"].dt.hour >= 18]

    power = evening[evening["Global_active_power"] > 6]

    group2 = power[
        (power["Sub_metering_2"] > power["Sub_metering_1"]) &
        (power["Sub_metering_2"] > power["Sub_metering_3"])
    ]

    half = len(group2) // 2

    first_half = group2.iloc[:half:3]
    second_half = group2.iloc[half::4]

    result = pd.concat([first_half, second_half])

    print("Кількість записів:", len(result))

    return result
```

```python
complex_selection(df)
```

### Завдання 7

Пронормувати вибраний датасет.

```python
def normalize_data(df):

    numeric_df = df.select_dtypes(include=["float64", "int64"])

    normalized = (numeric_df - numeric_df.min()) / (numeric_df.max() - numeric_df.min())

    return normalized
```

```python
normalized_df = normalize_data(df)
normalized_df.head()
```

### Завдання 8

Стандартизувати датасет.

```python
def standardize_data(df):

    numeric_df = df.select_dtypes(include=["float64", "int64"])

    standardized = (numeric_df - numeric_df.mean()) / numeric_df.std()

    return standardized
```

```python
standardized_df = standardize_data(df)
standardized_df.head()
```

### Завдання 9

Підрахувати коефіцієнти Пірсона та Спірмена.

```python
def correlations(df):

    numeric_df = df.select_dtypes(include=["float64", "int64"])

    pearson = numeric_df.corr(method="pearson")
    spearman = numeric_df.corr(method="spearman")

    print("Pearson:")
    print(pearson)

    print("\nSpearman:")
    print(spearman)
```

```python
correlations(df)
```

### Завдання 10

Провести One Hot Encoding категоріального атрибута.

```python
def one_hot_encoding(df):

    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

    df["Day"] = df["Date"].dt.day_name()

    encoded = pd.get_dummies(df, columns=["Day"])

    print("One Hot Encoding виконано")

    return encoded
```

```python
encoded_df = one_hot_encoding(df)
encoded_df.head()
```
